## Visualising Data part2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
```

## Create a data visualization with Pandas

### Bar plots

```python
df = pd.DataFrame(index = ['Atiya', 'Abbas', 'Cornelia','Stephanie', 'Monte'],
                 data = {'Apples':[20,10,40,20,50],
                        'Oranges':[45,40,25,19,33]})
```

```python
df
```

```python
#Bar plots use the index of the labels for the x axis and the column values as the bar heights. 
#color = ['.2','.7']
df.plot(kind='bar',figsize=(16,4)).legend(bbox_to_anchor=(0.9,1));
```

### KDE Plot

```python
#KDE plot ignores the index and uses the values of each column as the x axis and calculates a prob dens for the y values
df.plot(kind='kde',figsize=(16,4));
```

### Multi-variable plots

```python
#scatter plot is the only one that requires you to specify columns for the x and y values. If you wish to use 
#the index for a scatter plot, you will have to use the reset_index method to make it a column. 
#The other two plots use the index for the x axis and make a new set of lines/bars for every single numeric column:

fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(16,4))
fig.suptitle('Two variable Plots',size=20,y=1.02);
df.plot(kind='line', ax=ax1, title='Line plot')
df.plot(x='Apples', y='Oranges', kind='scatter',ax=ax2, title='Scatterplot')
df.plot(kind='bar', ax=ax3, title='Bar plot');
```

```python
#can specify which column to use
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
df.sort_values('Apples').plot(x='Apples',y='Oranges',kind='line',ax=ax1)
df.plot(x='Apples',y='Oranges',kind='bar',ax=ax2)
df.plot(x='Apples',kind='kde',ax=ax3)
```
